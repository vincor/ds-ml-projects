## Life Expectancy and GDP (MacroTrends&World Bank Data)
**In this project we will analyze data on GDP and life expectancy from the Macrotrends (The Premier Research Platform for Long Term Investors) and the World Bank to try and identify the relationship between the GDP and life expectancy of six countries from 2000 - 2022 years.**

Source of data:
    [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD)
    [Macrotrends](https://www.macrotrends.net/countries/ranking/life-expectancy)

During this project we will analyze, prepare, and plot data in order to answer questions in a meaningful way.




First we import necessary libraries. Then we load all data from the file 'all_data_full.csv' to the dataframe. We explore first 10 rows and information about the data structure.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
df = pd.read_csv("all_data_full.csv")
print(df.head(10))
print(df.info())
```

Our data structure looks good. There are no missing values. We can change the name of one column 'Life expectancy at birth (years)' for more convenient exploration. Let us change the name of this column to 'Life_Exp'. Also we change name of one of the countries 'United States of America' to 'USA'. To operate with billions of US dollars we change the value of the column 'GDP' by dividing it's value by 1bln.

```python
df = df.rename({"Life expectancy at birth (years)":"Life_Exp"}, axis = "columns")
df.loc[ df.Country == 'United States of America', 'Country'] = 'USA'
df.GDP = df.GDP/1000000000
df.head()
```

**Now we have our data well organized. It is without missing values, has adequate structure and variable names.**

```python
country_list = df.Country.unique()
print(country_list)
print(df.Year.unique())
```

**We are going to analyze years from 2000 to 2022 and six countries ('Chile' 'China' 'Germany' 'Mexico' 'USA'  and 'Zimbabwe').**

### 1. Exploratory analysis

We plot distributions of data, calculate means and standard deviation for each country and plot graphs with aggregated statistics.

```python
plt.figure(figsize=(14,8))
ax = plt.subplot()
sns.histplot(df.GDP, bins=15, color='#003388')
plt.xlabel("GDP in Billions of U.S. Dollars")    
plt.title('All countries')
plt.savefig('GDP_dist.png')
plt.show()
plt.clf()

plt.figure(figsize=(14,12))
for n in range(len(country_list)):
    ax = plt.subplot(3,2,n+1)
    sns.histplot(df[df.Country ==country_list[n]].GDP, bins=10, color='#0055ff')
    plt.title(country_list[n])
    plt.xlabel("GDP in Billions of U.S. Dollars")
    
    
plt.subplots_adjust(hspace=0.5)
plt.savefig('GDP_dist_countries.png')
plt.show()
plt.clf()
```

**We can see right skewness of the GDP distribution on the first graph for all countries. After detailed examination of distributions of each country we may conclude that overall GDP is very right skewed because of range of values of GDP of China and USA, that is much higher and wider than range for other countries.** 

```python
plt.figure(figsize=(14,8))
ax = plt.subplot()
sns.histplot(df.Life_Exp, bins=15, color='#007700')
plt.xlabel("Life expectancy at birth (years)")    
plt.title('All countries')
plt.show()
plt.clf()

plt.figure(figsize=(14,12))
for n in range(len(country_list)):
    #print(df[df.Country ==country_list[n]].head())
    ax = plt.subplot(3,2,n+1)
    sns.histplot(df[df.Country ==country_list[n]].Life_Exp, bins=10, color='#66ff66')
    plt.title(country_list[n])
    plt.xlabel("Life expectancy at birth (years)")
    
    
plt.subplots_adjust(hspace=0.5)
plt.show()
plt.clf()
```

**We can see very left skewness of the Life Expectancy distribution on the first graph for all countries. After detailed examination of distributions of each country we may conclude that the range of Life Expectancy for Zimbabwe is much lower than range of other countries. It is evident outlier in overall statistics.** 

```python
df_means = df.drop("Year", axis = 1).groupby("Country").mean().reset_index()
df_means
```

```python
df_std = df.drop("Year", axis = 1).groupby("Country").std().reset_index()
df_std
```

**We can notice significant values of standard deviation for GDP of China and USA. Also we can see a high value of the standard deviation for Life Expectancy in Zimbabwe. We will analyse reasons of that later while doing univariate analysis.** 

```python
plt.figure(figsize=(10,8))
ax = plt.subplot()
sns.barplot(data = df, x='Country', y='GDP', errorbar=('ci', 100), capsize=0.2, palette='colorblind')
plt.xlabel("Average GDP by countries in 2000-2022 years")
plt.savefig('Av_GDP.png')
plt.show()
plt.clf()

plt.figure(figsize=(10,8))
ax = plt.subplot()
sns.barplot(data = df, x='Country', y='Life_Exp', palette='mako', errorbar=('ci', 100), capsize=0.2)
plt.xlabel("Average Life Expectancy by countries in 2000-2022 years")
plt.ylabel('Years')
plt.ylim(40,100)
plt.savefig('Av_LEB.png')
plt.show()
plt.clf()
```

**On the graph of averages for GDP by countries we evidently can see outliers in values of GDP for USA and China and can find very wide ranges of GDP for these countries. Also average GDP of Chile is much lower and average GDP of Zimbabwe is almost unseen among other countries.
On the graph of averages for Life Expectancy by countries we may find low values and at the same time high range of this variable for Zimbabwe.** 

```python
plt.figure(figsize=(10,8))
ax = plt.subplot()
sns.boxplot(data = df, y='GDP', x='Country', palette='colorblind')
plt.xlabel("GDP by countries in 2000-2022 years")
plt.savefig('BP_GDP.png')
plt.show()
plt.clf()

plt.figure(figsize=(10,8))
ax = plt.subplot()
sns.boxplot(data = df, y='Life_Exp', x='Country', palette='mako')
plt.xlabel("Life Expectancy by countries in 2000-2022 years")
plt.ylabel('Years')
plt.savefig('BP_LEB.png')
plt.show()
plt.clf()
```

**The two previous graphs with boxplots for the variables GDP and Life Expectancy once again clearly confirm our conclusions regarding GDP and Life expectancy of the countries under consideration.**  

### 2. Univariate analysis

We plot line charts of GDP and Life Expectancy by years for all six countries and separately for each country.

```python
plt.figure(figsize=(14,8))
ax = plt.subplot()

sns.lineplot(data=df, x='Year', y='GDP', hue='Country', palette='colorblind')
plt.savefig('All_GDP.png')    
plt.show()
plt.clf()

plt.figure(figsize=(14,12))

for n in range(len(country_list)):
    ax = plt.subplot(3,2,n+1)
    sns.lineplot(data=df[df.Country ==country_list[n]], x='Year', y='GDP')
    plt.title(country_list[n])
    plt.xlabel('')
    
    
plt.subplots_adjust(hspace=0.4)
plt.savefig('All_GDP_Countries.png')
plt.show()
plt.clf()
```

**On the overall graph, we can clearly observe that the economies of China and the United States show impressive GDP growth compared to other countries. This explains the large values in standard deviations and the wide range of values.
The economies of Germany since 2008 and the economies of Chile and Mexico since 2012 have seriously slowed down GDP growth. The Zimbabwean economy, after stagnation until 2008, received good growth, but still the GDP is at a very low level compared to other countries.
Also on these charts, we can observe the events of the 2008-2010 crisis and the consequences of the 2020 coronavirus pandemic. It must be said that they were not reflected in the Chinese GDP chart and did not affect the US GDP chart much compared to strong shocks in other countries.**

```python
plt.figure(figsize=(14,8))
ax = plt.subplot()

sns.lineplot(data=df, x='Year', y='Life_Exp', hue='Country', palette='colorblind')
plt.ylabel("Life Expectancy (years)")
plt.savefig('All_LEB.png')    
plt.show()
plt.clf()

plt.figure(figsize=(14,12))

for n in range(len(country_list)):
    ax = plt.subplot(3,2,n+1)
    sns.lineplot(data=df[df.Country ==country_list[n]], x='Year', y='Life_Exp', color='#007700')
    plt.title(country_list[n])
    plt.xlabel('')
    plt.ylabel("Life Expectancy (years)")
    
    
plt.subplots_adjust(hspace=0.4)
plt.savefig('All_LEB_Countries.png')
plt.show()
plt.clf()
```

**On the overall graph of life expectancy by years, we see almost one high level for five countries, and very low, with a wide range of values for Zimbabwe.
Analyzing the graphs by country, we see a consistent increase in the values for Chile, China and Germany. On the US chart, we see growth until 2013, then we see an almost flat graph. Mexico, after the jump from 2000 to 2003, also shows more or less stable values at the same level, although not as high as other countries. And in a separate graph for Zimbabwe, we see very low values for Life Expectancy. Although it should be noted that since 2008 these values show an impressive increase from 45 to 62 years, this gives us a high standard deviation of this parameter for this African country.**

### 3. Bivariate analysis

In order to determine the presence of a correlation between parameters GDP and Life Expectancy, we will use scatterplot charts both for all data and for each country separately.

```python
plt.figure(figsize=(14,8))
ax = plt.subplot()
sns.scatterplot(data = df, y='Life_Exp', x='GDP', hue='Country', palette='colorblind')
plt.ylabel('Life Expectancy (years)')
plt.savefig('GDP_LEB_All.png')
plt.show()
plt.clf()

plt.figure(figsize=(14,12))

for n in range(len(country_list)):
    ax = plt.subplot(3,2,n+1)
    sns.regplot(data = df[df.Country ==country_list[n]], y='Life_Exp', x='GDP', color='#007700')
    plt.title(country_list[n])
    plt.ylabel("Life Expectancy (years)")
    
    
plt.subplots_adjust(hspace=0.4)
plt.savefig('GDP_LEB_Countries.png')
plt.show()
plt.clf()
```

**On the overall graph, due to the large spread of values for the GDP of China and the United States and the values of the Life Expectancy for Zimbabwe, the linear component for determining the correlation is not very noticeable.
But if we consider the same graphs separately for each country, the presence of a positive correlation between these parameters is obvious. This is most typical for Chile, China and Germany, where the spread of values relative to the linear trend is very small. Mexico stands apart, where the slope of the linear regression line is very small, which indicates a weak positive correlation.**
